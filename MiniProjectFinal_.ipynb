```python
import miniprojectmodule as mp
import tkinter as tk
import pandas as pd
def more_info(stock_symbol):
    print(f"Fetching more info for stock symbol: {stock_symbol}")

def on_enter_stock_symbol():
    stock_symbol = stock_symbol_entry.get()
    print(f"Basic info about the stock {stock_symbol}")
    mp.getbasic(stock_symbol)
    print("\n")
    
    
def on_more_info():
    stock_symbol = stock_symbol_entry.get()
    print(f"Advanced info about the stock {stock_symbol}")
    mp.get_more_info(stock_symbol)
    print("\n")

def on_chart():
    stock_symbol = stock_symbol_entry.get()
    print(f"Life Time Chart of the stock {stock_symbol}")
    mp.plot_historical_stock_data(stock_symbol,'1998-11-25','2023-12-02')
    print("\n")

def on_month():
    stock_symbol = stock_symbol_entry.get()
    print(f"Past Month Chart of the stock {stock_symbol}")
    mp.plot_onemonth_data(stock_symbol)
    print("\n")

def on_year():
    stock_symbol = stock_symbol_entry.get()
    print(f"Past Month Chart of the stock {stock_symbol}")
    mp.plot_oneyear_data(stock_symbol)
    print("\n")
    
def on_balance():
    stock_symbol = stock_symbol_entry.get()
    print(f"Balance sheeet of the stock {stock_symbol}(Values in Trillions)")
    print(mp.balance(stock_symbol))
    print("\n")
    
def on_visualize_sheet():
    stock_symbol = stock_symbol_entry.get()
    mp.visualize_balance_sheet(stock_symbol)
    
def on_csv():
    stock_symbol=stock_symbol_entry.get()
    df=mp.balance(stock_symbol)
    df.to_csv(stock_symbol+'balancesheet.csv')
    print(f"CSV sheet of {stock_symbol} is ready")
    
def on_exit():
    root.destroy()

root = tk.Tk()
root.title("Stock Information Interface")
root.geometry("600x600")

enter_stock_symbol_button = tk.Button(root, text="Get Basic Info: ", command=on_enter_stock_symbol)
enter_stock_symbol_button.pack(pady=10)

more_info_button = tk.Button(root, text="Get Advanced Info: ", command=on_more_info)
more_info_button.pack(pady=10)

chart_button = tk.Button(root, text="Life Time Chart: ", command=on_chart)
chart_button.pack(pady=10)

month_button = tk.Button(root, text="Previous Month Chart: ", command=on_month)
month_button.pack(pady=10)

year_button = tk.Button(root, text="Previous Year Chart: ", command=on_year)
year_button.pack(pady=10)

balance_button = tk.Button(root, text="Balance Sheet: ", command=on_balance)
balance_button.pack(pady=10)

csv_button = tk.Button(root, text="Convert Balance Sheet to CSV: ", command=on_csv)
csv_button.pack(pady=10)

visual_button = tk.Button(root, text="Visualize Balance Sheet: ", command=on_visualize_sheet)
visual_button.pack(pady=10)

exit_button = tk.Button(root, text="Exit", command=on_exit)
exit_button.pack(pady=10)

stock_symbol_label = tk.Label(root, text="Stock Symbol:")
stock_symbol_label.pack()

stock_symbol_entry = tk.Entry(root)
stock_symbol_entry.pack()


root.mainloop()
```
